```python

from google.colab import files
import io
uploaded = files.upload()


import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.decomposition import PCA

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (7,5)

if uploaded:
    fname = list(uploaded.keys())[0]
    df = pd.read_csv(io.BytesIO(uploaded[fname]))
    print("Loaded file:", fname, "| shape:", df.shape)
else:
    from sklearn.datasets import load_breast_cancer
    bc = load_breast_cancer(as_frame=True)
    df = bc.frame

    print("Using built-in breast cancer dataset | shape:", df.shape)

display(df.head())


possible_targets = ["diagnosis", "target", "class", "label", "Outcome"]
target_col = None
for c in possible_targets:
    if c in df.columns:
        target_col = c
        break
if target_col is None:
    target_col = df.columns[-1]  # fallback to last column
    print("No standard target name found — using last column as target:", target_col)
else:
    print("Detected target column:", target_col)


y_raw = df[target_col]
if y_raw.dtype == object:
    lowered = y_raw.astype(str).str.lower()
    if set(lowered.unique()) <= {"m","b","malignant","benign"}:
        y = lowered.map({"m":1,"malignant":1,"b":0,"benign":0}).astype(int)
    elif set(lowered.unique()) <= {"yes","no"}:
        y = lowered.map({"yes":1,"no":0}).astype(int)
    else:

        y = pd.to_numeric(y_raw, errors="coerce")
        if y.isna().any():
            raise ValueError("Target column contains strings I cannot map automatically. Please convert to 0/1.")
        y = y.astype(int)
else:
    y = y_raw.astype(int)


X = df.drop(columns=[target_col], errors="ignore").copy()

for col in ["id","Id","patient_id","PatientID","Unnamed: 0"]:
    if col in X.columns:
        X = X.drop(columns=[col])
X = X.select_dtypes(include=[np.number])
X = X.fillna(X.median())

print("\nFinal shapes -> X:", X.shape, " y:", y.shape)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)


svc_linear = SVC(kernel="linear", probability=True, random_state=42)
svc_linear.fit(X_train_sc, y_train)
y_pred_lin = svc_linear.predict(X_test_sc)
y_prob_lin = svc_linear.predict_proba(X_test_sc)[:,1]

print("\n--- Linear SVM ---")
print("Accuracy:", round(accuracy_score(y_test, y_pred_lin), 3))
print(classification_report(y_test, y_pred_lin))
print("ROC-AUC:", round(roc_auc_score(y_test, y_prob_lin), 3))


cm_lin = confusion_matrix(y_test, y_pred_lin)
plt.figure()
sns.heatmap(cm_lin, annot=True, fmt="d", cmap="Blues")
plt.title("Linear SVM - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()


svc_rbf = SVC(kernel="rbf", probability=True, random_state=42)
svc_rbf.fit(X_train_sc, y_train)
y_pred_rbf = svc_rbf.predict(X_test_sc)
y_prob_rbf = svc_rbf.predict_proba(X_test_sc)[:,1]

print("\n--- RBF SVM ---")
print("Accuracy:", round(accuracy_score(y_test, y_pred_rbf), 3))
print(classification_report(y_test, y_pred_rbf))
print("ROC-AUC:", round(roc_auc_score(y_test, y_prob_rbf), 3))

cm_rbf = confusion_matrix(y_test, y_pred_rbf)
plt.figure()
sns.heatmap(cm_rbf, annot=True, fmt="d", cmap="Blues")
plt.title("RBF SVM - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()


fpr1, tpr1, _ = roc_curve(y_test, y_prob_lin)
fpr2, tpr2, _ = roc_curve(y_test, y_prob_rbf)
plt.figure()
plt.plot(fpr1, tpr1, label=f"Linear (AUC={roc_auc_score(y_test,y_prob_lin):.3f})")
plt.plot(fpr2, tpr2, label=f"RBF (AUC={roc_auc_score(y_test,y_prob_rbf):.3f})")
plt.plot([0,1],[0,1],"k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend()
plt.show()

if X_train_sc.shape[1] >= 2:
    pca = PCA(n_components=2, random_state=42)
    Xtr2 = pca.fit_transform(X_train_sc)
    Xte2 = pca.transform(X_test_sc)
    clf_plot = SVC(kernel="rbf", probability=True, random_state=42).fit(Xtr2, y_train)


    x_min, x_max = Xtr2[:,0].min()-1, Xtr2[:,0].max()+1
    y_min, y_max = Xtr2[:,1].min()-1, Xtr2[:,1].max()+1
    xx, yy = np.meshgrid(np.linspace(x_min,x_max,200), np.linspace(y_min,y_max,200))
    Z = clf_plot.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    plt.figure(figsize=(7,6))
    plt.contourf(xx, yy, Z, alpha=0.2)
    plt.scatter(Xtr2[:,0], Xtr2[:,1], c=y_train, edgecolor="k", cmap="viridis", alpha=0.8)
    plt.title("RBF SVM Decision Regions (PCA 2D projection)")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.show()

print("\nDone — models trained and evaluated. Save notebook and upload to your repo for submission.")
```
